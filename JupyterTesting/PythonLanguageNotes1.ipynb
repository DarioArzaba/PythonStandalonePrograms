```python
# Regresa el (Valor Absoluto) de un numero entero o flotante. Si es un numero complejo se regresa su magnitud.
absoluteNumber = abs(-1.24)
print("Abs:", absoluteNumber)
print()

# Regresa (True) solo si todos los elementos de un iterable son verdaderos o si el iterable esta vacio.
everyElementTrue1 = all( [True, True, True] )
everyElementTrue2 = all( [True, False, True] )
print("All1:", everyElementTrue1)
print("All2:", everyElementTrue2)
print()

# Regresa (True) si cualquier elemento del iterable es verdadero.
anyElementTrue1 =  any( [False, False, False] )
anyElementTrue2 = any( [False, True, False] )
print("Any1:", anyElementTrue1)
print("Any2:", anyElementTrue2)
print()

# Regresa un (String) que representa un objeto como en Python 2.
# Es decir que es similar a repr() pero escapa los caracteres no-ASCII en forma de Unicode.
asciiRepresentObject = ascii('Pythön')
print("Ascii:", asciiRepresentObject)
print()

# Regresa un (String Binario) con el prefijo 0b de un numero entero.
# Para evitar el uso del prefijo, usar la funcion format().
binarioNumero = bin(3)
binarioNumeroFormato = format(34, 'b')
binarioNumeroFormatoDos = format(34, '#b')
print("Bin1:", binarioNumero)
print("Bin2:", binarioNumeroFormato)
print("Bin3:", binarioNumeroFormatoDos)
print()

# Constructor de la clase Bool, subclase de Int.
# Regresa un (Booleano) a partir de un objeto. 
# Todo objeto es verdadero a menos que defina un metodo __bool__() que regrese falso.
# Tambien puede ser falso si su metodo __len__() regresa cero. 
# Constantes y números falsos: None, False, 0, 0.0, 0j, Decimal(0), Fraction(0,1).
# Secuencias y coleccioines falsas: '', (), [], {}, set(), range(0)
isItBool = bool(0)
print("Bool:", isItBool)
print()

# Activa el debugger al llamar a sys.breakpointhook() ligado a pdb.set_trace() 
# Funcion de conveniencia que pasa sus argumentos a breakpointhook()
#breakpoint(*args, **kws)

# Constructor de la clase ByteArray,
# Regresa un (ByteArray) a partir de un objeto.
# Un bytearray es una secuencia mutable de bytes (enteros del 0 al 256).
#     Un (string) necesita un encoding y usa str.encode().
#     Un (entero) genera un bytearray de ese tamaño con null bytes.
#     Un (iterable) de enteros genera bytearray en ese rango.
byteArray1 = bytearray("Hello", 'utf-8')
byteArray2 = bytearray(5)
byteArray3 = bytearray([2,3,6])
print("ByteArray1:", byteArray1)
print("ByteArray2:", byteArray2)
print("ByteArray3:", byteArray3)
print()

# Constructor de la clase Bytes.
# Regresa varios (Bytes) a partir de un objeto.
# Los Bytes son una secuencia inmutable de bytes (enteros del 0 al 256).
#     Un (string) necesita un encoding y usa str.encode().
#     Un (entero) genera tantos null bytes.
#     Un (iterable) de enteros genera bytes en ese rango.
bytes1 = bytes("Hello", 'utf-8')
bytes2 = bytes(5)
bytes3 = bytes([2,3,6])
print("Bytes1:", bytes1)
print("Bytes2:", bytes2)
print("Bytes3:", bytes3)
print()

# Regresa (False) si el objeto del argumento no se puede llamar.
# Si regresa True esto no quiere decir que el objeto se pueda llamar.
# Todas las clases se pueden llamar, al hacerlo creamos una nueva instancia.
# Toda instancia que tenga el metodo __call__() se puede llamar.
integerNumber = 1
def testMethod():
    print("Success")
class Foo1:
    pass
class Foo2:
    def __call__(self):
        print("Instance Call : Success")
InstanceOfFoo2 = Foo2()
callable1 = callable(integerNumber)
callable2 = callable(testMethod)
callable3 = callable(Foo1)
callable4 = callable(InstanceOfFoo2)
print("Callable1:", callable1)
print("Callable2:", callable2)
print("Callable3:", callable3)
print("Callable4:", callable4)
# Llamar a la instancia ejecuta su metodo __call__()
InstanceOfFoo2() 
print()

# Regresa un caracter (String) utilizando un entero como codigo unicode.
# Para obtener el entero de un string se utiliza ord()
# El rango de enteros es de 0 a 1,114,111 (0x10FFFF en x16).
chr1 = chr(97)
chr2 = chr(8364)
print("Chr1:", chr1)
print("Chr2:", chr2)
print()

# Regresa un metodo de clase a partir de un metodo normal.
# Los metodos de clase reciven a la clase como su primer argumento implicito (cls)
# A diferencia de los estaticos, los de clase pueden acceder de forma limitada a su clase.
# Se recomienda utilizar el decorador @classmethod.
class Foo3:
    @classmethod
    def testFunction(cls):
        print("Class method: ", end = '')
        cls.printFunction()
    def printFunction():
        print("Success.")
Foo3.testFunction()       
print()

# Regresa un (Objeto Codigo) a partir de un source compilado.
# El source puede ser un string, un byte string o un AST.
# El modo especifica como se ejecutara el objeto codigo:
#     Exec es para ejecutar multiples sentencias.
#     Eval es para ejecutar una sola expresion.
#     Single es para ejecutar una sola expresion interactiva.
sourceCode = 'a = 5\nb=6\nsum2=a+b\nprint("Compile (5+6):", sum2)'
compile1 = compile(sourceCode, 'sumProgramName', 'exec')
exec(compile1)
print()

# Constructor de la clase Complex
# Regresa un numero (Complex) compuesto por un valor real y uno imaginario (j).
# Se puede utilizar un string (sin espacios) o dos valores numericos.
# Se pueden crear numeros complejos como si fueran literales.
complex1 = complex("2+3j")
complex2 = complex(2, -3)
complex3 = complex(2)
complex4 = 2+3j
print("Complex1:", complex1)
print("Complex2:", complex2)
print("Complex3:", complex3)
print("Complex4:", complex4)
print()

# Delattr borra un atributo de un objeto. Es un complemento de setattr()
# Tambien se pueden eliminar atributos usando el operador (del)
class Foo4:
    atribute1 = 10
    atribute2 = 20
    atribute3 = 30
delattr(Foo4, "atribute3")
del Foo4.atribute2
InstanceofFoo4 = Foo4()
print("Delattr:", InstanceofFoo4.atribute1)
print()

# Constructor de la clase Dict
# Regresa un (Diccionario). Otros contenedores son list, set y tuple
# Se pueden contruir mediante argumentos, iterables, mapas o un literal.
dictArguments = dict(a=1, b=2)
dictIterables = dict([("c",3),("d",4)]) 
dictMapping = dict( {'e': 5} , f=6 )
dictLiteral = {"g":7, "h":8}
print('Dict1:', dictArguments)
print('Dict2:', dictIterables)
print('Dict3:', dictMapping)
print('Dict4:', dictLiteral)
print()

# Regresa una (Lista) con los nombres del scope local.
# Con un objeto regresa la lista de sus atributos.
#      Si el objeto tiene un metodo __dir__(self) este es llamado para la lista.
#      Si el objeto no tiene __dir__ se recopilan los nombres del atributo __dict__.
# El metodo se creo para generar los nombres más utiles, no necesariamente exclusiva.
class Foo5:
    def __dir__(self):
        return ["area", "perimeter", "location"]
InstanceofFoo5 = Foo5()
print('Dir: ', dir(InstanceofFoo5))
print()

# Regresa un (Tuple) con el cociente y sobrante de una division entera.
# Utiliza dos numeros no complejos como argumentos.
divmodInteger = divmod(8,3)
divmodFloat = divmod(7.5, 2.5)
print('Divmod1:', divmodInteger)
print('Divmod2:', divmodFloat)
print()

# Regresa un objeto enumerate a partir de un iterable.
# El metodo __next__() del iterable es el que otorga el index de cada elemento.
enumerateIterable = ["One", "Two", "Three", "Four"]
enumerateObject = enumerate(enumerateIterable, start=1)
print("Enumerate:", list(enumerateObject))
print()

# El argumento se evalua como una expresion de Python.
# Regresa el resultado de la expresion evaluada. Utiliza el scope que lo rodea.
# Si la expresion solo contiene literales se recomienda usar ast.literal_eval()
evalInteger = 5
evalExpression = eval("evalInteger+6")
print("Eval (5+6):", evalExpression)
print()

# El argumento se evalua de forma dinamica, debe ser un string u objeto codigo.
# Utiliza el scope que lo rodea. 
execProgram = 'a = 5\nb=6\nprint("Exec (5+6) =", a+b)'
exec(execProgram)
print()

# Regresa un (Iterator) a partir de elementos de un iterable 
# El iterable puede ser una secuencia, un contenedor u otro iterador
# Se aplica una funcion para cada elemento y solo se agrega al iterator si la funcion genera True
# Este metodo es equivalete al generador (item for item in iterable if function(item))
filterText = ['a', 'b', 'd', 'e', 'i', 'j', 'o']
def filterVowels(source):
    vowels = ['a', 'e', 'i', 'o', 'u']
    if(source in vowels):
        return True
    else:
        return False
filteredSource = filter(filterVowels, filterText)    
filterResult=""
for vowelFound in filteredSource:
    filterResult+=vowelFound
print("Filter:", filterResult)
print()

# Constructor de la clase Float
# Regresa un numero flotante a partir de otro numero o de un string.
# Para muchos objetos este metodo llama a la funcion __float__()
floatingNumber = "+1.34"
print("Float:", floatingNumber)
print()

# Regresa la representacion formateada de un valor 
# Para generar la representacion llama a Object.__format__(format_spec)
# Los tipos de datos incluidos usan Format Specification Mini Language como format_spec.
# Una version modificada y más avanzada de este metodo es utilizada para strings.
#    fillChar, align, sign, width, comma, precision, type
#    AnyChar, [<,>,=,^], [+,-, ], [int], [,], [.int], [b,c,d,e,f,g,n,o,s,x,X,%]
formatInteger = format(123, "*>+7,d")
formatFloat = format(123.4567898, "^-09.3f")
formatBinary = format(12, "b")
print("Format1:", formatInteger)
print("Format2:", formatFloat)
print("Format3:", formatBinary)
print()

# Constructor de la clase Frozenset
# Regresa un objeto (frozenset) a partir de elementos de un iterable.
# Es la versión inmutable de un (Set)
frozensetIntegers = frozenset([2,3,6])
print("Frozenset:", frozensetIntegers)
print()

# Regresa un atributo de un objeto.
# Equivalente a la notacion punto de un objeto.
class Foo6:
    atribute1 = 10
InstanceofFoo6 = Foo6()
print("Getattr:", getattr(InstanceofFoo6, "atribute1"))
print()

# Regresa un (dict) con la tabla de simbolos globales.
# Corresponde siempre al modulo en donde se define la funcion o clase.
# En Jupyter todo el documento pertenece al diccionario del modulo.
# globalsDict = globals()
# print("Globals:", globalsDict)

# Regresa un atributo de un objeto. Es un complemento de setattr()
# Equivalente a la notacion punto de un objeto.
class Foo7:
    atribute1 = 10
InstanceofFoo7 = Foo7()
print("Hasattr:", hasattr(InstanceofFoo7, "atribute1"))
print()

# Regresa el (hash) de un objeto inmutable o si esta definido en __hash__().
# Los hash son numeros enteros que se usan para comparar rapidamente las llaves de un diccionario.
# Por ejemplo Python los usa para comparar la igualdad de numeros aunque sean de distinto tipo (int-float)
hashOneInt = hash(1)
hashOneFloat = hash(1.0)
isOneIntEqualOneFloat = (hashOneInt == hashOneFloat)
print("Hash:", hashOneInt, "==", hashOneFloat, " = ", isOneIntEqualOneFloat)
print()

# Regresa el sistema de ayuda incluido. Para uso en el shell interactivo.
# Su argumento permite buscar modulos, funciones, clases, metodos, keyword o tema en la ayuda.
# help("keyword")

# Regresa el (hex) con el prefijo 0x de un numero entero.
# Si el objeto no es un numero entero necesita definir __index__() para que regrese un entero.
hexNumber = hex(255)
hexNumberUpper = format(255, "X")
hexNumberPrefix = format(255, "#x")
hexFloat = float.hex(255.123)
print("Hex1:", hexNumber)
print("Hex2:", hexNumberUpper)
print("Hex3:", hexNumberPrefix)
print("Hex4:", hexFloat)
print()

# Regresa el address (Id) en memoria de un objeto.
# Este entero es siempre unico y constante para cada objeto mientras este vivo.
# Si el objeto es eliminado un nuevo objeto podria obtener el mismo Id.
class Foo7:
    x = 5
InstanceofFoo7 = Foo7()
idOfInstance = id(InstanceofFoo7)
idOfNumber = id(1)
print("Id1:", idOfInstance)
print("Id2:", idOfNumber)
print()

# Si tiene un argumento string lo escribe al Standard Output.
# Despues lee una linea del Standard Input, la convierte a un string y la regresa.
# Si en la linea se lee un EOF, se regresa una excepcion EOFError.
# El modulo (readline) modifica el comportamiento de input()
# inputHello = input("Enter a Message: ")
# print("Input:", inputHello)

# Constructor de la clase Int
# Regresa un objeto numero entero a partir de un numero o string.
# Si el argumento es otro objeto simplemente llama a su metodo __int__() o __trunc__().
# Tambien se puede seleccionar la base numerica del entero.
intNumber = int(123)
intHex = int("C", 16)
print("Int1:", intNumber)
print("Int2:", intHex)
print()

# Regresa (True) si el objeto es una instancia de la clase o subclases.
# Se compara con un tuple con el identificador de una o más clases.
isInstanceVerification = isinstance(10, (int))
print("Isinstance:", isInstanceVerification)
print()

# Regresa (True) si la clase es una subclase de otra clase.
# Se considera a una clase como una subclase de si misma.
isSubclassVerification = issubclass(bool, int)
print("Issubclass:", isSubclassVerification)
print()

# Regresa un (iterator) a partir de un objeto.
# Debe ser una coleccion iterable como un set o tuple.
# Si no es una coleccion iterable debe ser una secuencia que:
#     Implementa __iter__() y __next__()
#     o Implementa __getitem__() con argumentos enteros iniciando en 0
# Si hay un (sentinel) como argumento el objeto debe ser callable.
# En el caso de un sentinel se llamara al objeto sin argumentos usando __next__().
# Si el objeto regresa el valor del sentinel se generara la excepcion StopIteration.
vowelsList = ['a', 'e', 'i', 'o', 'u']
vowelsIter = iter(vowelsList)
next(vowelsIter)
print("Iter:", next(vowelsIter))
print()
 
# Regresa el tamaño (numero de elementos) en un objeto.
# El objeto puede ser una secuencia o una coleccion.
# String, bytes, tuples, lists, ranges. Dictionarys, sets, frozensets.
lenghtSequence = len("Hello")
lenghtCollection = len({"a":1, "b":2})
print("Len1:", lenghtSequence)
print("Len2:", lenghtCollection)
print()

# Constructor de la clase List
# Regresa una (List) a partir de un iterable.
# El iterable puede ser una secuencia, coleccion o iterator.
# La lista por si sola es una secuencia mutable de elementos.
listVowelSet = {'a', 'e', 'i', 'o', 'u'}
listVowel = list(listVowelSet)
print("List:", listVowel)
print()

# Regresa un (dict) con la tabla de simbolos del scope local.
# Si se llama desde un modulo locals() y globals() son equivalentes.
# Puede que los cambios a locals() no se apliquen por eso solo es para propositos de lectura.
def randomLocalsFunc():
    x = 3
    return locals()
localsDict = randomLocalsFunc()
print("Locals:", localsDict)
print()

# Un iterator aplica una funcion a cada elemento de un iterable.
# Regresa un (Map Object) con los resultados.
# Generalmente se utilizan con funciones lambda
def mappingFunction(a,b):
    return (a+b)
mapInputElementsA = (1,2,3,4)
mapInputElementsB = (5,3,2,1)
mapResults1 = map(mappingFunction, mapInputElementsA, mapInputElementsB)
mapResults2 = map(lambda n: (n*n), mapInputElementsA)
print("Map1:", list(mapResults1))
print("Map2:", list(mapResults2))
print()

# Regresa el maximo (max) o minumo (min) de un iterable o de una lista de argumentos.
# Se puede incluir una funcion para aplicar a los elementos antes de compararlos.
# Se puede incluir un valor por defecto en caso de que la lista este vacia.
maxFromArguments = max(1,2,3,4,5)
maxFromIterable = max([5, 2, 7, 3, 4, 9])
def minSumFunction(number):
    sum1 = 0
    while (number):
        # Mientras tanto va sumando los sobrantes
        sum1 = sum1 + (number % 10)
        # Se detiene cuando la divicion entre 10 llega a dar 1
        number = int(number/10) 
    # Se genera la suma de sobrantes
    return sum1
minFromFunction = min(102,132,122,13,242, key=minSumFunction)
print("Max1:", maxFromArguments)
print("Max2:", maxFromIterable)
print("Min:", minFromFunction)
print()

# Regresa un (Memory View) a partir de un objeto.
# El protocolo buffer de Python es la representacion interna de objetos.
# A cada objeto se le asigna una memory array o buffer en la memoria.
# Cada vez que modificamos un objeto Python crea una copia completa del buffer del objeto.
# Para acceder y modificar una seccion del buffer en vez de copiarlo por completo usamos memoryviews
memoryViewsArray = bytearray("abcde", "utf-8")
memoryView = memoryview(memoryViewsArray)
print("Memoryview:", list(memoryView))
print()

# Regresa el siguiente (__next__()) elemento de un iterator.
# Puede regresar un valor por defecto cuando se llege al final del iterator.
nextIteratorList = [5, "cat"]
nextIterator = iter(nextIteratorList)
nextIteratorCalls = []
for i in range(3):
    nextIteratorCalls.append(next(nextIterator, "-1"))
print("Next:", nextIteratorCalls)
print()

# Regresa el (Oct) con el prefijo 0o de un numero entero.
# Si el objeto no es un numero entero necesita definir __index__() para que regrese un entero.
octNumber = oct(255)
octNumberUpper = format(255, "o")
octNumberPrefix = format(255, "#o")
print("Oct1:", octNumber)
print("Oct2:", octNumberUpper)
print("Oct3:", octNumberPrefix)
print()

# Constructor de la clase Object
# Esta clase especial es la clase de la que se derivan todas las demas.
# Las instancias de esta clase no tienen un atributo diccionario __dict__.
primaryObjectTest = object()
print("Object:", dir(primaryObjectTest)[0:3])
print()

# Abre un archivo y crea un (File Object) a partir de el.
# Si el archivo no se puede abrir se genera una excepcion OSError
# Sus argumentos son:
#     File: Es el path u objeto path, absoluto o relativo, del archivo.
#     Mode: esstring que especifica como abrir el archivo, por defecto es rt = read text.
#          r(read), w(overwrite), x(create), a(append), b(binary), t(text), +(disk file update)
#          En modo texto si el encoding no es especifica se asume de locale.getpreferredencoding(0)
#     Buffering: Especifica cuantos bytes acepta el buffer, 0 para apagarlo, 1 para lineas.
#          Mas de 1 en buffering para especificar manualmente el numero de bytes (auto: 4k - 8k)
#     Errors: Handling de errores generados por el encoding o decoding en modo texto.
#          strict(ValueError), ignore(ignorar), replace(?), surrogateescape(unknown codepoints)
#          xmlcharrefreplace(&#nnn XML char refs), backslashreplace(escape char).
# import os
# desktopPath = os.path.expanduser("~/Desktop/")
# with open(desktopPath+"test.txt", "x") as f:
#     print('Hello World!', file=f)

# Regresa un entero que representa un codigo unicode en base a un caracter
# Esta funcion es el complemento de chr()
ordCodePoint = ord('€')
print("Ord:", ordCodePoint)
print()

# Regresa un numero (x) elevado a (y).
# Si existe un tercer argumento (z) se aplica su modulo.
numberPowerFunction = pow(2, 4, 3)
print("Pow:", numberPowerFunction)
print()

# Imprime sus argumentos al archivo (Stream de Texto) seleccionado.
# Los argumentos son separados (sep) y lineas finalizan con (end)
# Se puede seleccionar otro archivo (file) para logging llamando el metodo write(str).
# Si write(str) no esta definido se utiliza sys.stdout(). No se puede imprimir a binarios.
# Si el output se guarda en un buffer lo determina el archivo pero el argumento flush lo elimina.
print("Print", "Meta", sep=": ", end=None)
print()

# Funcion que establece el getter, setter y deleter de un atributo.
# Se recomienda utilizar su decorador @property
# Los metodos se llaman cuando se intenta obtener, cambiar o borrar la propiedad.
class Foo8:
    def __init__(self, name):
        self.name = name
    def getName(self):
        return self._name + " :)"
    def setName(self, newName):
        newName = newName + "!"
        self._name = newName
    def delName(self):
        del self._name
    name = property(getName, setName, delName, "Name Property")
InstanceOfFoo8 = Foo8("Old Name")
InstanceOfFoo8.name = "New Name"
print("Property:", InstanceOfFoo8.name)
print()

# Constructor de secuencias inmutables (Range) de numeros enteros
# Sus argumentos son el inicio, final y step de la secuencia.
rangeNumericTest = range(2, 19, 3)
print("Range:", rangeNumericTest)
print()

# Regresa un (String) con la representacion de un objeto
# Se genera el nombre, address y atributos del objeto.
# Se puede controlar lo que se genera con el metodo __repr__()
representationObject = repr(str(123))
print("Repr:", representationObject)
print()

# Regresa un (Iterator) con sus elementos invertidos.
# La secuencia debe definir el metodo __reversed__().
reversedIterator = reversed("abcdef")
print("Reversed:", list(reversedIterator))
print()

# Regresa un numero entero redondeado a cierta precision.
# Generalmente llama al metodo __round__()
roundedNumberInteger = round(123.501)
roundedNumberFloat = round(123.487, 2)
print("Round1:", roundedNumberInteger)
print("Round2:", roundedNumberFloat)
print()

# Constructor de la clase Set
# Genera un nuevo (set) con elementos de un iterable.
setInstance = set(("a", "b", "c", "d", "e"))
print("Set:", setInstance)
print()

# Agrega un atributo a un objeto
# Los argumentos son la instancia, el nombre del atributo y un valor.
# Es el complemento de getattr(). Puede referirse a un atributo ya existente.
class Foo9:
    pass
InstanceOfFoo9 = Foo9()
InstanceOfFoo9.atributo1 = 10
setattr(InstanceOfFoo9, "atributo2", 20)
print("Setattr:", InstanceOfFoo9.atributo1, InstanceOfFoo9.atributo2)
print()

# Regresa un (Objeto Slice) que representa un set de indices mediante un Range.
# No tienen funcionalidad más que para el funcionamiento interno de otras funciones.
# En concreto se utiliza cuando se realiza el Extended Indexing Syntax a[i:f:s]
# Para generar un iterator a partir de una operacion slice usar itertools.islice().
characterRange = "Hello"
sliceObject = slice(1, 3, 1)
print("Slice:", characterRange[sliceObject])
print()

# Regresa una (Lista) organizada ascendente con los elementos de un iterable.
# Con sus argumentos se puede aplicar una funcion a cada elemento o revertirlos todos.
# Sorted es estable, es decir no cambia el orden relativo de elementos que sean iguales.
sortedVowelList = ['e', 'a', 'u', 'o', 'i']
print("Sorted1:", sorted(sortedVowelList))
print("Sorted2:", sorted(sortedVowelList, reverse=True))
print()

# Regresa un metodo estatico de una clase a partir de uno normal.
# Los metodos estaticos no tienen argumento implicito, son autonomos.
# Se recomienda utilizar el decorador @staticmethod.
class Foo10:
    @staticmethod
    def testFunction():
        print("Static: ", "Success")
Foo10.testFunction()
print()

# Constructor de la clase String
# Regresa un String a partir de un objeto.
stringFromObject = str(123.5)
print("String:", stringFromObject)
print()

# Regresa el resultado de aplicar la operacion de suma a un iterable.
# Generalmente los elementos son numeros y se suman de izquierda a derecha.
# Para sumar strings se recomienda "".join(sequence) y para float math.fsum()
# Para concatenar iterables se recomienda itertools.chain()
sumInputRange = range(1, 10)
sumRangeOfNumbers = sum(sumInputRange)
print("Sum:", sumRangeOfNumbers)
print()

# Regresa un objeto proxy que delega sus metodos a una clase padre.
# Este objeto permite acceder a metodos heredados que han sido sobreescritos.
class Foo11Parent:
    def testFunc(self, n):
        # super(Foo11, self).testFunc(n)
        n *= 3
        return n
class Foo11(Foo11Parent):
    def testFunc(self, n):
        # super(Foo11, self).testFunc(n)
        n += 1
        super().testFunc(n)
InstanceOfFoo11 = Foo11()
bothFunctions = InstanceOfFoo11.testFunc(2)
print("Super:", bothFunctions)
print()

# Constructor de la clase Tuple
# Regresa una secuencia inmutable de tipo Tuple a partir de un iterable
tupleFromList = tuple([1,2,3,4,5])
print("Tuple:", tupleFromList)
print()

# Regresa un (Objeto Type) que representa la clase de otro objeto
# El objeto type normalmente es generado por el atributo __class__
# Se recomienda isinstance() para comparar tipos de datos por considerar subclases.
# Un uso concreto es para la creacion dinamica de clases.
class typeClass1:
    x = 2
InstanceOfTypeClass1 = typeClass1()
InstanceOfTypeClass2 = type("typeClass2", (object,), dict(x=2))
print("Type1:", InstanceOfTypeClass1.x, InstanceOfTypeClass2.x)
print("Type2:", type(InstanceOfTypeClass1))
print()

# Regresa el atributo __dict__ de un modulo, clase u objeto.
# Este atributo guarda un diccionario o mapa interno con los atributos del modulo, clase u objeto.
# Las clases usan types.MappingProxyType para prevenir la edicion directa de su __dict__
# Sin argumentos es el equivalente mutable de locals().
class Foo12:
    def __init__(self):
        self.a = 5
        self.b = 10
InstanceOfFoo12 = Foo12()
print("Vars:", vars(InstanceOfFoo12))
print()

# Regresa un (iterator) de tuples mediante los elementos de varios iterables.
# Si los iterables son de distinta longitud de usa el de menor tamaño
# Si se quiere usar la de mayor tamaño usar itertools.zip_longest(). 
# Se pueden generar los iterables originales (unzip) con el operador(*)
unzippedListFirst = [1, 2, 3, 7]
unzippedListSecond = [4, 5, 6]
zippedList = zip(unzippedListFirst, unzippedListSecond)
zippedListCopy = zip(unzippedListFirst, unzippedListSecond)
unzippedListFirst2, unzippedListSecond2 = zip(*zippedListCopy)
print("Zip1:", list(zippedList))
print("Zip2:", list(unzippedListFirst2))
print("Zip3:", list(unzippedListSecond2))
print()

# Funcion especial llamada de forma interna al usar el keyword import
# Permite modificar la forma por defecto en la que se importan modulos.
# Sin embargo se recomienda usar importlib.import_module() para importaciones personalizadas.
# El bytecode que obtenemos de las importaciones por defecto es:
#     import spam ->
#     spam = __import__('spam', globals(), locals(), [], 0)
#     import spam.ham ->
#     spam = __import__('spam.ham', globals(), locals(), [], 0)
#     from spam.ham import eggs, sausage as saus ->
#     _temp = __import__('spam.ham', globals(), locals(), ['eggs', 'sausage'], 0)
#     eggs = _temp.eggs
#     saus = _temp.sausage
```
